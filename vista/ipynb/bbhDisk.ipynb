## Uniform Grid

### Scaling

```julia
let patt_dirs = [
        (r"Cow_L12_G32-N\d+-MPI\d+_r0000", "G32"),
        (r"Cow_L12_G64-N\d+-MPI\d+_r0000", "G64"),
        (r"Cow_L12_G128-N\d+-MPI\d+_r0000", "G128"),
        (r"Cow_L12_G256-N\d+-MPI\d+_r0000", "G256"),
    ]
    (dats, labs) = Misc.load_avgs(patt_dirs, "../data/bbhDisk"; option="TotalComputeTime") 
    #
    plt = plot(ylabel="M/day")
    for (i, dat) in enumerate(dats)
        plot!(plt, dat[1], dat[2], label=labs[i],
              marker=:circle, xscale=:log10, yscale=:log10)
        println(dat[2])
    end
    display(plt)
end
```

```julia
let patt_dirs = [
        (r"Cow_L12_G32-N\d+-MPI\d+_r0000", "G32"),
        (r"Cow_L12_G64-N\d+-MPI\d+_r0000", "G64"),
        (r"Cow_L12_G128-N\d+-MPI\d+_r0000", "G128"),
        (r"Cow_L12_G256-N\d+-MPI\d+_r0000", "G256"),
    ]
    (dats, labs) = Misc.load_avgs(patt_dirs, "../data/bbhDisk", option="ZcsPerSecond") 
    #
    plt = plot(ylabel="Zcs/sec")
    for (i, dat) in enumerate(dats)
        plot!(plt, dat[1], dat[2], label=labs[i],
              marker=:circle, xscale=:log10, yscale=:log10)
        println(dat[2])
    end
    display(plt)
end
```

### Speed

```julia
let dirs = [
        ("Cow_L12_G256-N4-MPI4_r0000/stdout.txt", "N4"),
        ("Cow_L12_G256-N8-MPI8_r0000/stdout.txt", "N8"),
        ("Cow_L12_G256-N16-MPI16_r0000/stdout.txt", "N16"),
        ("Cow_L12_G256-N32-MPI32_r0000/stdout.txt", "N32"),
    ]
    (dats, labs) = Misc.load_data(dirs, "../data/bbhDisk", "TotalComputeTime")
    #
    plt = plot()
    for (i, dat) in enumerate(dats)
        plot!(plt, dat[1], dat[3], label=labs[i])
    end
    display(plt)
end
```
